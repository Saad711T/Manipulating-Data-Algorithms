```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(0,9)
y = (25,33,41,53,59,70,78,86,98)
```

```python
plt.scatter(x,y)

z = np.polyfit(x,y,1)
p = np.poly1d(z)

plt.title('Graph')
plt.plot(x,p(x), 'g-')
plt.show()
```

```python
print(np.poly1d(p))
```

```python
print(np.poly1d(p(6)))
```
